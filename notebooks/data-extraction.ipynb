```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm
from urllib.request import urlretrieve 
from urllib.request import urlopen, Request
```

```python
import requests

# api url from ECI
url = "https://register.eci.ec.europa.eu/core/api/register/details/2024/000007"

headers = {
    "User-Agent": "Mozilla/5.0"
}

res = requests.get(url, headers=headers)
data = res.json()

print(data.keys())
```

```python
df = pd.json_normalize(data)
df.T.head(20)
```

```python
signature_data = data.get('submission', {}).get('entry', [])

df = pd.DataFrame(signature_data)

print(df[['countryCodeType', 'total']].head())
```
